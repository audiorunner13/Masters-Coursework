```python
import matplotlib.pyplot as plt
```

```python
import networkx as nx
```

```python
neighbors={'Andre':['Fernando', 'Beverly', 'Diane','Carol'], 'Beverly':['Andre','Diane','Garth','Ed'],
           'Diane':['Fernando', 'Beverly', 'Andre', 'Garth','Ed','Carol'],'Carol':['Andre','Diane','Fernando'], 
           'Fernando':['Carol','Andre','Diane', 'Garth', 'Heather'], 'Ed':['Beverly', 'Diane', 'Garth'], 
           'Garth':['Ed','Beverly','Diane','Fernando','Heather'], 'Heather':['Garth','Fernando','Ike'], 
           'Ike':['Heather', 'Jane'], 'Jane':['Ike']}
```

```python
print(neighbors)
```

```python
G = nx.Graph()
```

```python
for neighbor in neighbors:
    G.add_node(neighbor)
    for n in neighbors[neighbor]:
        G.add_edge(neighbor, n)
        
print(G.nodes)
```

```python
nx.draw_networkx(G, node_color = 'r', edge_color = 'b', with_labels=True)
```
